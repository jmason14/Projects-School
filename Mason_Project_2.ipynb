# Gases, Fractals and Root Finding
Ideal gas law is the equation of a hypothetical state of gas. The ideal gas equation is,

$$ PV = nRT $$

where $P = pressure$, $V = volume$, $n =$ $number$ $of$ $molecules$ , $R =$ $gas$ $constant$ and $T = temperature$. The ideal gas law is thought to be a good approximation of the behavior of gases; however, it has several limitations. Thus, many scientists have attempted to develop newer more accurate ways of predicting behavior of real gases.

__Part 1 - Critical Points of the van der Waals equation__
<br>
<br>
The van der Waals equation(1857) is named after Dutch theoretical physicist, Johannes Diderik van der Waals. Van der Waal's equation is based on reasons why real gases do not follow ideal gas law. When he was studying ideal gas law, van der Waal noticed that there is pressure not accounted for in the ideal gas equation. For real gases there exists an attractive interaction force that affects these gases. Additionally, van der Waal noticed that the ideal gas equation did not account for all the volume that a real gas takes up. To account for the added pressure of attraction and volume occupied, van der Waal devised his own equation that tries harder to capture the properties of real gas. The van der Waal equation is written as,

$$ (P + \frac{an^2}{V^2})(V - b) = nRT $$

Here, $(V-b)$ represents the available volume insider of a container and $\frac{an^2}{V^2}$ is the interaction term. So what can be done with the van der Waal equation? The van der Waal equation attempts to predicts phase transitions more accurately than ideal gas law. Phase transitions tend to occur at critical phenomena. In other words, when constructing a model considering $P$, $V$ and $T$ the critical points of such model will represent a phase transition. To take a closer look at this behavior, we'll use the reduced gas equation. The reduced gas equation is given by,

$$ (p_r + \frac{3}{v_r^2})(v_r - \frac{1}{3}) = \frac{8}{3}T_r $$

If we solve for $p_r$ we get,

$$ p_r(v_r, T_r) = \frac{8T_r}{3v_r - 1} - \frac{3}{v_r^2}. $$

This reduced gas equation can be used to create isotherms for van der Waal gas.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def vdw(v_r = 1, T_r = 0.8):
    """
    Function uses the van der Waal's reduced equation.
    
    Input:
    v_r - reduced Volume
    T_r - reduced Temperature
    
    Output:
    p_r - reduced Pressure
    
    """
    return (8*T_r)/(3*v_r-1)-(3/v_r**2)
```

```python
T_r1 = 0.5
T_r2 = 0.6
T_r3 = 0.75
T_r4 = 0.8
T_r5 = 0.9
T_r6 = 1
T_r7 = 1.1
T_r8 = 1.25
T_r9 = 1.3
T_r10 = 1.4
T_r11 = 1.5
v_r1 = np.linspace(0, 4, 1000)

# Plot isotherms using the van der Waal equation

fig, axes = plt.subplots(figsize = (12,8))
axes.plot(v_r1, vdw(v_r1, T_r1), label = "$0.5$")
axes.plot(v_r1, vdw(v_r1, T_r2), label = "$0.6$")
axes.plot(v_r1, vdw(v_r1, T_r3), label = "$0.75$")
axes.plot(v_r1, vdw(v_r1, T_r4), label = "$0.8$")
axes.plot(v_r1, vdw(v_r1, T_r5), label = "$0.9$")
axes.plot(v_r1, vdw(v_r1, T_r6), label = "$1$")
axes.plot(v_r1, vdw(v_r1, T_r7), label = "$1.1$")
axes.plot(v_r1, vdw(v_r1, T_r8), label = "$1.25$")
axes.plot(v_r1, vdw(v_r1, T_r9), label = "$1.3$")
axes.plot(v_r1, vdw(v_r1, T_r10), label = "$1.4$")
axes.plot(v_r1, vdw(v_r1, T_r11), label = "$1.5$")
axes.set_xlabel("Reduced Volume ($v_r$)", fontsize = 15)
axes.set_ylabel("Reduced Pressure ($p_r$)", fontsize = 15)
axes.set_title("$p_r$ vs $v_r$", fontsize = 15)
axes.set_ylim([-2, 5])
axes.legend()
```

It may not be easily noticeable a first glance, but there is an inflection point at $p_r = 1$, $v_r = 1$, $T_r = 1$. Upon calculation, we will find that the critical temperature is $T_r = 1$. To find an inflection point, we set the second derivative equal to zero like so,

$$ f''(x) = 0. $$

Here, $f(x)$ will equal the expanded reduced form of the van der Waal's equation given as,

$$ v_r^3 - \frac{1}{3}(1 + \frac{8T_r}{p_r})v_r^2 + \frac{3}{p_r}v_r - \frac{1}{p_r}. $$

So,

$$ f(x) = v_r^3 - \frac{1}{3}(1 + \frac{8T_r}{p_r})v_r^2 + \frac{3}{p_r}v_r - \frac{1}{p_r}, $$

when $p_r = 1$, $T_r = 1$,

$$ v_r^3 - \frac{1}{3}9v_r^2 + \frac{3}{1}v_r - 1, $$

it follows,

$$ f'(x) = 3v_r^2 - 6v_r + 3, $$

$$ f''(x) = 6v_r - 6. $$

Set $f''(x)$ equal to zero,

$$ 6v_r - 6 = 0, $$

$$ 6v_r = 6, $$

$$ v_r = 1. $$

Thus, we have confirmed mathematically that the inflection point lies at $p_r = 1$, $v_r = 1$, $T_r = 1$. Now let's zoom in on the plot of isotherms and analyze them visually.

In terms of phase transition, the critical temperature is the where the liquid is in the process of turning into a gas. At the critical temperature liquids and gases can coexist. We notice that $p_r = 0.4$ intersects the $T_r = 0.8$ isotherm three times! This suggests that there exists three roots to the equation. These roots can be found using the Bisection Method. The bisection method is is a root-finding method that repeatedly bisects an interval, then selects a subinterval in which a root must lie for further processing. The bisection method is very good at finding roots, but can be slow. An important part to the bisection method is the intermediate value theorum, which it uses to help locate the roots. Let's use the bisection method to find the roots when $p_r = 0.4$ and $T_r = 0.8$. To do this we will need the expanded reduced form of the van der Waal's equation. This is given as

$$ v_r^3 - \frac{1}{3}(1 + \frac{8T_r}{p_r})v_r^2 + \frac{3}{p_r}v_r - \frac{1}{p_r} = 0. $$

```python
def vroot(v_r = 1, T_r = 0.8, p_r = 0.4):
    return v_r**3 - (1/3)*(1 + 8*T_r/p_r)*v_r**2 + 3*v_r/p_r - 1/p_r
```

```python
def BisectionFunc(f = vroot, a = .4, b = 1.0, tolerance = 1e-7):
    """
    Returns the root of the input function and the number of iterations as a pair of values using the bisection method.
    
    Keywords:
    f - function whose root I want to calculate
    a = start of the interval
    b - end of the interval
    tolerance - number of sig figs I need
    """
    
    signa = np.sign(f(a))
    signb = np.sign(f(b))
    numIter = 0
    
    while(b-a > tolerance):
        c = (a+b)/2
        print("Best Guess: {}".format(c))
        signc = np.sign(f(c))
        
        if signc == 0:
            return c
        elif signc == signa:
            a = c
        else:
            b = c
        numIter += 1
        
    return c, numIter
```

The bisection method is a reliable method of finding roots. However, one can see how slow this method can become. Even for a computer, the number of iterations pile up leading to a longer wait period. This could become an issue with more complex functions. Is there a faster method? Let's look at another example of a root finding algorithm; the Newton-Raphson Method. The Newton-Raphson method is a method for finding successively better approximations to the roots of a function. As opposed to the bisection method, Newton-Raphson method uses derivatives to create approximations. One starts with an initial guess as to where a root lies and then the function is approximated by a tangent line. Then the x-intercept of this tangent line is computed, which will typically be a better guess of where the real root lies. The method is then iterated in order to find the root. The iteration formula is

$$ x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}. $$

A large downside to Newton-Raphson method is that it does not always converge. If a function is not well behaved, the method could overshoot and diverge. Nevertheless, let's use the Newton-Raphson method so it can be compared with the bisection method used previously.

```python
def vroot2(v_r = 1, T_r = 0.8, p_r = 0.4):
    return v_r**3 - (1/3)*(1 + 8*T_r/p_r)*v_r**2 + 3*v_r/p_r - 1/p_r, 3*v_r**2 - (2/3)*(1+8*T_r/p_r)*v_r + 3/p_r
```

```python
def newtonMethod(fx = vroot2, xInit = .5, tolerance = 1e-8):
    """
    Function returns the root of the input function f and the number of iterations it takes using Newton's method
    
    Keywords:
    
    f- function which returns f and df as a pair of values
    xInit - intial guess
    tolerance - number of sig figs I want to know reliably
    
    output - pair of values, first is the root, second is number of iterations
    
    """
    # Initialize the variables
    
    step = 0
    f, df = fx(xInit)
    dx = -f/df
    x = xInit
    
    # Run the iterations in the algorithm
    
    while np.abs(dx) > tolerance:
        print("Best guess of root = {} at iteration (:3d)".format(x, step)) # (:3d) is a print limiter
        f, df = fx(x)
        dx = -f/df
        x += dx
        step += 1
    
    # Return output    
    return x, step
```

__Part 2 - Complex Numbers and Fractals in the Complex Plane__

A complex number is a number that can be expressed in the form $x + jy$, where $x$ and $y$ are real numbers. Consider the equation $x^2 = -1$. There is no real number that satisfies this equation, thus the need for the creation of complex numbers. Complex numbers provide solutions to equations such as $x^2 = -1$ that have no solution in the real numbers. Complex number, $i$, is the solution to the equation $x^2 = -1$. $i$ is referred to as an imaginary number. Knowing this, consider solving for the roots of the equation

$$ z^3 = 1. $$

The equation can be written in the form, 

$$ x + jy $$

where $j = \sqrt{-1}$. Thus, equation $z^3 = 1$ can be rewritten,

$$ z^3 = 1 + 0j $$ 

There is a certain strategy to solving an equation like this. A French mathematician by the name of Abraham de Moivre developed a theorum cleverly named de Moivre's theorum. De Moivre's theorum states that for any complex number, $x$, and integer $n$ it holds that

$$ [cos(x) - jsin(x)]^n = cos(nx) + jsin(nx). $$

de Moivre's theorum can be used to solve for the roots of $z^3 = 1$.
<br>
<br>
Using de Moivre's theorum,

$$ z^3 = 1 $$

$$ z^3 = cos(2\pi{k}) + j*sin(2\pi{k}) $$

$$ z^3 = cos(3\frac{2\pi}{3}k) + j*sin(3\frac{2\pi}{3}k) $$

$$ z^3 = [cos(\frac{2\pi}{3}k) + j*sin(\frac{2\pi}{3}k)]^3 $$

Then,

$$ z = cos(\frac{2\pi}{3}k) + j*sin(\frac{2\pi}{3}k) $$

for k = 0, 1, 2.
<br>
<br>
Plugging k into the equation the roots are,

$$ z_0 = cos(0) + j*sin(0) = 1 + 0j $$

$$ z_1 = cos(\frac{2\pi}{3}) + j*sin(\frac{2\pi}{3}) = -0.5 + \frac{\sqrt{3}}{2}j $$

$$ z_2 = cos(\frac{4\pi}{3}) + j*sin(\frac{4\pi}{3}) = -0.5 - \frac{\sqrt{3}}{2}j $$

Now when solving using python, we know what the results should look like.

```python
import numpy as np
import matplotlib.pyplot as plt
```

The image above is a fractal. Fractals are infinitely complex patterns and exhibit self-similarity. The self-similarity of this image can be seen if we zoom in.

```python
# Zoom in
x = np.linspace(-.5, .5, numpts)
y = np.linspace(-.5j, .5j, numpts)
X, Y = np.meshgrid(x,y)
Z = X + Y
niters = 100

for i in range(niters):
    Z = Z - (Z**3-1)/(3*Z**2)

tol = .01

root1 = np.abs(Z - roots[0]) < tol
root2 = np.abs(Z - roots[1]) < tol
root3 = np.abs(Z - roots[2]) < tol

img = np.zeros((numpts, numpts, 3))
img[:,:,0] = root1 # Red
img[:,:,1] = root2 # Blue
img[:,:,2] = root3 # Green
fig = plt.figure(figsize = (8,8))
plt.imshow(img)
plt.show()
```
